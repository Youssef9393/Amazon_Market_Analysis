```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = pd.read_parquet('../data/raw/amzn.parquet')
df.index = pd.to_datetime(df.index)
```

```python
df['Volume'].plot(title="Amazon volume evolution between 2020 et 2025 en Europe",figsize=(10,5))
plt.grid(True)
```

```python
# calculation of daily returns
df['returns'] = df['Close'].pct_change()
```

```python
display(df.head(5))
```

```python
display(df['returns'].describe())
```

```python
plt.figure(figsize=(10,3))
sns.histplot(df['returns'].dropna(), bins=50, kde=True, color='purple')
plt.title("Distribution of daily returns of Amazon ")
plt.xlabel("Daily Returns")
plt.ylabel("frequence")
plt.show()
```

```python
import numpy as np
window = 30  # 30 jours
df['rolling_mean'] = df['Close'].rolling(window).mean()
df['rolling_std'] = df['Close'].rolling(window).std()
df['vol_annualized'] = df['rolling_std'] * np.sqrt(252)
```
